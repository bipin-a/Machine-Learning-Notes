```python
import numpy as np
import matplotlib.pyplot as plt
from math import ceil, floor
```

- Randomly generate 10,000 integers uniformly distributed in [1, 200] and store them in an array named x.
- Sample 100 numbers randomly from x, and find their mean
    -  Now repeat the sampling 50,000 times and put the mean values in a new array named meen.
     




```python
N=10000
low = 1
high = 200
X = np.random.randint(low=low,high=high+1, size=N)
```

```python
sample_iterations = 50000
meen = np.empty(sample_iterations)
for i in range(sample_iterations):
    sample_size = 100
    sample = np.random.choice(X,size=sample_size)
    meen[i] = np.mean(sample)
```

Create a frequency plot - one which shows how many times each mean appeared. 
The mean values in the horizontal axis and the frequency in the vertical axis. 


```python
# Approximately dedicate 1 bin for each unique integer
bin_num = ceil(max(np.unique(meen))) - floor(min(np.unique(meen)))
m = round(np.mean(meen),2)
std_dev = round(np.std(meen),2)

fig, ax = plt.subplots(figsize=(9, 6))
n, edges, patches = ax.hist(meen,bins=bin_num)
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Frequency Count')
ax.set_title('Histogram of Samples')
ax.text(85, 3000, f'$\mu$:{m} , $\sigma$: {std_dev}')
ax.grid(which='major', alpha=0.5)
```

```python
mask_range = (edges <=115) & (edges >=85)
```

```python
range_index = np.where(mask_range)
n_range = n[range_index]
edge_range = edges[mask_range]
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
b = ax.bar(edge_range,n_range)
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Frequency Count')
ax.set_title('Histogram of Samples Between 85 and 115')
ax.grid(which='major', alpha=0.5)
```
